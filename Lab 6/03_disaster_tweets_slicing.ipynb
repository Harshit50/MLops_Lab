```python
import pandas as pd
from utils import load_disaster_dataset
from snorkel.slicing import slicing_function, PandasSFApplier

# 1. Load the data (we only need the test dataset for slicing)
_, _, df_test, _ = load_disaster_dataset()

# 2. Write rules to find specific subsets of data
@slicing_function()
def slice_is_retweet(x):
    # Identifies if a tweet is a retweet
    return x.text.startswith("RT ") or " RT " in x.text

@slicing_function()
def slice_has_link(x):
    # Identifies if a tweet contains a web link
    return "http" in x.text.lower()

# Group our slicing functions together
sfs = [slice_is_retweet, slice_has_link]

# 3. Apply the slices using the correct Pandas applier
applier = PandasSFApplier(sfs)
S_test = applier.apply(df=df_test)

# 4. View a summary of how many tweets fell into each slice
print("--- Slice Analysis Summary ---")
print(f"Total Test Tweets: {len(df_test)}")

print(f"Tweets in 'Retweet' slice: {S_test['slice_is_retweet'].sum()}")
print(f"Tweets in 'Has Link' slice:  {S_test['slice_has_link'].sum()}")

# 5. View the actual text of the isolated "Retweet" data
rt_slice_df = df_test[S_test['slice_is_retweet'] == 1]

print("\n--- Sample of Isolated Retweets ---")
display(rt_slice_df[['text', 'label']].head())
```
